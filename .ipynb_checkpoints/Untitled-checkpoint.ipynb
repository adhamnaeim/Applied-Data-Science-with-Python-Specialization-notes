# Regex manipulation

```python
import re
```

```python
string = 'Today is a bad day!'

#first we will use re.search which search for a string entry inside your desired string.

#compared to the traditional method of using in
if "good" in string:
    print("good good")
else:
    print('-_-')

#re.search method.
if re.search("good", string):
    print("indeed it was wonderful!")
else:
    print(":(")
```

```python
#spliting and findall
string = "adham is coding. adham is working. adham is exercising"
print(re.split("adham",string)) #re.splits the array and removes the searching string 
print(string.split("adham"))  #string.split works in the same fashion where it removes the searching string
print(re.findall("adham",string)) #findall appends all instances of the particular string into an array
```

```python
#starting with (^) and ending with ($)
print(string)
print(len(string))
print(string[44:54])
print(re.search("^adham",string)) #if this returns an object, then the string infact begins with the string, indexing in the range of the span
print(re.search("exercising$",string)) #if this returns an object, then the string infact ends with the string, indexing in the range of the span
```

```python
#patterns and character classes
#lets say we have a string that contains the grades of students and we want to do some data acquisition
string = "CADBABADAAAAAAADBCBADCBAEAAACBADCBAD"

#will print each time an "A" is find
print(re.findall("A",string), len(re.findall("A",string)) ) 

```

```python
#will print each time an "B" is find
print(re.findall("B",string), len(re.findall("B",string)))

```

```python
#will print each time an "AB" is find CONSECUTIVELY
print(re.findall("AB",string), len(re.findall("AB",string))) 

```

```python
#it is basically the code of lines 4 and 5 combined in one line.
print(re.findall("[AB]",string) , len(re.findall("[AB]",string))) 

```

```python
#the - indicated "between" two elements
print(re.findall("[AB]",string) , len(re.findall("[D][A-C]",string))) 

```

```python
#the "|" indicates "or" function where it will search for d followed by a or b or c
print(re.findall("[AB]",string) , len(re.findall("[D][A|B|C]",string))) 

```

```python
#Not using pipe operator "|" yields the same results i think
print(re.findall("[AB]",string) , len(re.findall("D[ABC]",string)))
```

```python
# using "^" inside the set operator "[]" while gather everything that does not include the string
print(re.findall("[^A]",string) , len(re.findall("[^A]",string)))
```

```python
#confusing part! what if you use the caret sign "^"inside and outside the set operator?
print(re.findall("^[^A]",string) , len(re.findall("^[^A]",string)))
#it will output if the string starts with anything other than A
```

```python
#the regex quantifiers "{}" are used to find the streak for the expression we are matching
#we put the expression outside and the minimum and maximum number we want to find this expression
print(re.findall("A{2,10}",string) , len(re.findall("A{2,10}",string)))
```

```python
#the regex quantifier is stupid, if you leave empty spaces inside the operator, 
#it will output an empty array without throwing any errors!
#                    \/
print(re.findall("A{2, 10}",string) , len(re.findall("A{2, 10}",string)))
```

```python
#but the regex quantifier has also an interesting feature 
#where you can have only one input inside which will be both min and max!
print(re.findall("A{2}",string) , len(re.findall("A{2}",string)))
```

```python
#the infamous "?"(0 or 1) "*" (0 or more) and "+" (1 or more)
string = "ABBD"
```

```python
with open("ferpa.txt","r") as file:
    wiki = file.read()

wiki
```

```python
#finds the last 100char that are capital or small letters
#before the string "[edit]"
re.findall("[a-zA-Z]{1,100}\[edit\]",wiki) 
```

```python
#\w imdicates all letters and digits, \s matches any whitespace
re.findall("[\w]{1,100}\[edit\]",wiki) 
```

```python
#we can use the asterisk to replace the {1,100}
re.findall("[\w]*\[edit\]",wiki) 
```

```python
#now lets clean it up and delete the [edit] at the end
for title in re.findall('[\w ]*\[edit\]', wiki):
    print(title.split('[')[0])
```

```python
#use grouping "()" to group patterns.  
#we can group the [\w ]* and the \[edit\] independently
re.findall(('([\w ]*)(\[edit\])'),wiki)
```

```python
#groups are helpful when trying to slice or take our certain elements that occur in a pattern
#using its position
#for example we can use re.finditer to return an output with the found groups in tuples.
for item in re.finditer(('([\w ]*)(\[edit\])'),wiki):
    #here we print all the elements in all the groups
    print(item.group(0))
    #here we printed all the elements in the first group which we are interested in
    print(item.group(1))
    #and finally we print all the elements in the 2nd group
    print(item.group(2))
    
```

```python
#we can actually name each group to create a dictionary, where the name of the group acts as our key.
#we add a name to a group by adding ?P<the group name> inside the beginning of the grouping parameter ()

for item in re.finditer(('(?P<titles>[\w ]*)(?P<edit_link>\[edit\])'),wiki):
    print(item['titles'])
    print(item.groupdict()['titles'])
```

```python
#we can also use ?= to denote that a certain group is neglectable
for item in re.finditer(('(?P<titles>[\w ]+)(?=\[edit\])'),wiki):
    print(item)
```
